```python
import joblib
from mecab import MeCab
from tqdm import tqdm
import pandas as pd
import numpy as np 
import glob

folder_list = ['.\\color\*','.\\outfit\*','.\\price\*','.\\quality\*','.\\status\*','.\\texture\*','.\\thick\*']
model_list=[]
for folder in folder_list:
    file_list = glob.glob(folder +'*.pkl')  
    model_list.append(file_list)
```

```python
color_molder,color_col,color_scaler,color_vetorizer=joblib.load(model_list[0][0]),joblib.load(model_list[0][1]),joblib.load(model_list[0][2]),joblib.load(model_list[0][3])

outfit_molder,outfit_col,outfit_scaler,outfit_vetorizer=joblib.load(model_list[1][0]),joblib.load(model_list[1][1]),joblib.load(model_list[1][2]),joblib.load(model_list[1][3])

price_molder,price_col,price_scaler,price_vetorizer=joblib.load(model_list[2][0]),joblib.load(model_list[2][1]),joblib.load(model_list[2][2]),joblib.load(model_list[2][3])

quality_molder,quality_col,quality_scaler,quality_vetorizer=joblib.load(model_list[3][0]),joblib.load(model_list[3][1]),joblib.load(model_list[3][2]),joblib.load(model_list[3][3])

status_molder,status_col,status_scaler,status_vetorizer=joblib.load(model_list[4][0]),joblib.load(model_list[4][1]),joblib.load(model_list[4][2]),joblib.load(model_list[4][3])

texture_molder,texture_col,texture_scaler,texture_vetorizer=joblib.load(model_list[5][0]),joblib.load(model_list[5][1]),joblib.load(model_list[5][2]),joblib.load(model_list[5][3])

thick_molder,thick_col,thick_scaler,thick_vetorizer=joblib.load(model_list[6][0]),joblib.load(model_list[6][1]),joblib.load(model_list[6][2]),joblib.load(model_list[6][3])
```

```python

filter=['NNG','MAG','EC','VA','VA+EF','VV+ETM','NNB+JKB','VCP+EC','VCP','MAG+JX','VCN']

def tokenizer(data):
    tokenizer = MeCab()
    
    list = []
    
    for text in tqdm(data["리뷰"]):
        prah = []  
        lst = tokenizer.pos(text)  
        for word, pos in lst:
            if pos in filter:
                prah.append(word)  
        list.append(' '.join(prah))  
    
    return pd.DataFrame({'tokens':list})
```

```python
sample_data=pd.DataFrame([['이옷 색감이 너무 좋아와요 화면이랑 완전 똑같아서 맘에 듭니다 ㅎㅎ '],
                          ['가격은 좋은데 좀 옷이 많이 까끌까끌 그려서 살짝 아쉽네요 하지만 이정도 퀄리티에 이정도가격이면 만족합니다'],
                          ['옷 너무 좋아요 특히나 색감 화면 나온 완전 똑같아서 맘에 듬요'],
                          ['옷 자체는 너무 이쁘고 핏도 이뻐요, 근데 약품냄새가 쫌 나서 빨아서 입어야할거 같아요.. 학교갈때 입어도 좋을거 같아요'],
                          ['핏도 예쁘고 부들부들하고 탄탄해요!!색은 생각했던 것보다 약간 노란끼가 돌긴 하는 것 같아용! 그래두 넘 예쁩니댜..🤍'],
                          ['기장이 길지도 짧지도 않은 애매~~ 한 길이예요 키가 커서 넣어입거나 접어입는 게 이쁠 것 같아요 넥라인 완전 작고 목 짧은 사람은 비추합니닷! 생각보다 완전 빵빵해서 슈퍼 어깡쟁이^^ 꼬임디테일 귀엽지만 깡마른 사람이 입으면 더 잘 어울릴 것 같아용']],columns=['리뷰'])

sample_data=tokenizer(sample_data)
sample_data
```

```python
def color_kawrg(data):
    x=color_vetorizer.transform(data['tokens']).toarray()
    x=color_scaler.transform(x)    
    x=x[:,color_col]
    return color_molder.predict(x)

def outfit_kawrg(data):
    x=outfit_vetorizer.transform(data['tokens']).toarray()
    x=outfit_scaler.transform(x)    
    x=x[:,outfit_col]
    return outfit_molder.predict(x)

def price_kawrg(data):
    x=price_vetorizer.transform(data['tokens']).toarray()
    x=price_scaler.transform(x)    
    x=x[:,price_col]
    return price_molder.predict(x)

def quality_kawrg(data):
    x=quality_vetorizer.transform(data['tokens']).toarray()
    x=quality_scaler.transform(x)    
    x=x[:,quality_col]
    return quality_molder.predict(x)

def status_kawrg(data):
    x=status_vetorizer.transform(data['tokens']).toarray()
    x=status_scaler.transform(x)    
    x=x[:,status_col]
    return status_molder.predict(x)

def texture_kawrg(data):
    x=texture_vetorizer.transform(data['tokens']).toarray()
    x=texture_scaler.transform(x)    
    x=x[:,texture_col]
    return texture_molder.predict(x)

def thick_kawrg(data):
    x=thick_vetorizer.transform(data['tokens']).toarray()
    x=thick_scaler.transform(x)    
    x=x[:,thick_col]
    return thick_molder.predict(x)
```

```python
sample_data['색_리뷰']=color_kawrg(sample_data)
sample_data['핏_리뷰']=outfit_kawrg(sample_data)
sample_data['가격_리뷰']=price_kawrg(sample_data)
sample_data['퀄리티_리뷰']=quality_kawrg(sample_data)
sample_data['제품상태_리뷰']=status_kawrg(sample_data)
sample_data['재질_리뷰']=texture_kawrg(sample_data)
sample_data['두께_리뷰']=thick_kawrg(sample_data)
```
